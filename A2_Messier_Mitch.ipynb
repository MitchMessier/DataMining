# Task IV

```python
car_auction.head()
```

```python
car_auction['over_80k_miles'] = car_auction['VehOdo'] > 80000
car_auction.head()
```

```python
filtered_df = car_auction[car_auction['over_80k_miles']]
filtered_df = filtered_df.sort_values(by='VehOdo', ascending=False)
filtered_df.head()
```

# Task V

```python
fig, ax = plt.subplots(figsize=(10,6))
plt.boxplot(filtered_df['VehOdo'])
plt.show()
```

This is a box plot of of vehicle miles. We can see by this box plot that the average or mean miles on a car is around 85,000 miles. We can also see that the upper quartile is up to 100000 and above while the lower quartile is 80000 and below.

```python
plt.figure(figsize=(10, 6))
filtered_df['Size'].value_counts().plot(kind='bar')
plt.title('Bar Plot of Vehicle Sizes')
plt.xlabel('Size')
plt.ylabel('Count')
plt.show()
```

This is a barplot displaying all of the auctioned cars by the count within each category. Just looking at the data we can see that medium, large, and medium SUV's are the most popular. Medium cars clearly pull away with the most cars and sports cars make up the least.

```python
plt.figure(figsize=(8, 6))
plt.hist(filtered_df['WarrantyCost'], bins=20, edgecolor='black')
plt.title('Histogram of Warranty Costs')
plt.xlabel('Warranty Cost')
plt.ylabel('Frequency')
plt.show()
```

Looking at this histogram, we've split it up into multiple bins where we can see where most warrenty costs fall into. We can see that it is slighlty a bell curve where most warrenty costs fall in the 1000 to 2000 range while others fall outside of this range but less frequently.
